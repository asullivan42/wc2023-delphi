```python
import numpy as np
import pandas as pd
```

```python
WC_U20 = pd.read_csv('../Datasets/FBRef/U20.csv', encoding='ISO-8859-1')
```

```python
WC_U20_long = pd.melt(WC_U20, id_vars = 'Team', var_name = 'Year', value_name = 'Rank')
```

```python
WC_U20_long = WC_U20_long[WC_U20_long['Year'] != 'Total']
```

```python
WC_U20_long = WC_U20_long[WC_U20_long['Rank'] != '\x95']
```

```python
WC_U20_long
```

```python
WC_U20_long.Rank.value_counts()
```

```python
WC_U20_long = WC_U20_long[WC_U20_long['Rank'] != '×']
```

```python
WC_U20_long.Rank.value_counts()
```

```python
WC_U20_long['Team'] = WC_U20_long['Team'].replace('United States','USA')
```
