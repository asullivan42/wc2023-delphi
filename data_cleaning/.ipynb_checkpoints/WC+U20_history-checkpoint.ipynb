```python
import numpy as np
import pandas as pd
```

```python
WC_U20 = pd.read_csv('../Datasets/FBRef/U20.csv', encoding='ISO-8859-1')
```

```python
WC_U20.head()
```

```python
WC_U20.dtypes
```

```python
WC_U20.iloc[0,0]
```

```python
WC_U20.drop(0, inplace = True)
```

```python
WC_U20.head()
```

```python
WC_U20['Team'] = WC_U20['Team'].str.strip() 
```

```python
WC_U20_long = pd.melt(WC_U20, id_vars = 'Team', var_name = 'Year', value_name = 'Rank')
```

```python
WC_U20_long = WC_U20_long[WC_U20_long['Year'] != 'Total']
```

```python
WC_U20_long = WC_U20_long[WC_U20_long['Rank'] != '\x95']
```

```python
WC_U20_long
```

```python
WC_U20_long.Rank.value_counts()
```

```python
WC_U20_long = WC_U20_long[WC_U20_long['Rank'] != '×']
```

```python
WC_U20_long.Rank.value_counts()
```

```python
WC_U20_long['Team'] = WC_U20_long['Team'].replace({'United States':'USA'})
```

```python
WC_U20_long['Rank'] = WC_U20_long['Rank'].replace({'GS':'Group'})
```

```python
WC_U20_long.head()
```

```python
WC_U20_long.to_csv('../Datasets/cleaned_data/WC_U20.csv') 
```

```python
WC = pd.read_csv('../Datasets/cleaned_data/WC_history.csv', index_col = 0)
```

```python
WC = WC[['Rk', 'Squad', 'Year']]
```

```python
WC.columns = ['Rank','Team', 'Year']
```

```python
WC = WC.reindex(columns = ['Team', 'Year', 'Rank'])
```

```python
WC['Rank'] = WC['Rank'].replace({'GR':'Group', '1': '1st', '2':'2nd', '3':'3rd', '4':'4th'})
```

```python
WC.head(20)
```

```python
WC_concat = pd.concat([WC.assign(source='WC'), WC_U20_long.assign(source='WC_U20')])
```

```python
WC_concat
```

```python
WC_concat.Team.sort_values().unique()
```

```python
WC_concat.Team = WC_concat.Team.replace({'ChinaPR':'China', 'ChineseTaipei':'Chinese Taipei', 'CostaRica':'Costa Rica', 
                        'NewZealand':'New Zealand', 'KoreaDPR':'North Korea', 'KoreaRep': 'South Korea', 'SouthAfrica':'South Africa'})
```

```python
#save for tableau
WC_concat.to_csv('../Datasets/cleaned_data/WC_concat.csv') 
```
