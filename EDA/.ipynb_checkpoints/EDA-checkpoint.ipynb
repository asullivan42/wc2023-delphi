```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
```

```python
sns.set()
```

```python
rating16_22 = pd.read_csv('../Datasets/cleaned_data/FIFA16-22_rating_new.csv')
```

```python
rating16_22.head()
```

```python
rating16_22.shape
```

```python
len(rating16_22['short_name'].unique())
```

```python
len(rating16_22['long_name'].unique())
```

```python
rating16_22_avg = rating16_22.groupby(by = ['short_name', 'long_name'], as_index = False).mean()
```

```python
duplicateRows = rating16_22_avg[rating16_22_avg.short_name.duplicated(keep = False)]
duplicateRows
```

```python
rating16_22_avg.shape
```

```python
duplicateRows = rating16_22_avg[rating16_22_avg.long_name.duplicated(keep = False)]
duplicateRows
```

```python
rating16_22_avg = rating16_22.groupby(by = ['long_name'], as_index = False).mean()
```

```python
rating16_22_avg.shape
```

```python
add_nation = rating16_22[['long_name','nationality_name']].drop_duplicates('long_name')
```

```python
rating16_22_new = pd.merge(rating16_22_avg, add_nation, on = 'long_name', how = 'left')
```

```python
rating16_22_new.head()
```

```python
rating16_22_new.shape
```

```python
rating16_22_new.rename(columns = {'nationality_name':'nationality'},inplace = True)
```

```python
means = rating16_22_new.groupby(['nationality'])['overall'].mean().sort_values()
```

```python
plt.subplots(figsize=(12, 6))
sns.stripplot(x='nationality', y='overall', data=rating16_22_new, order = means.index, palette = 'crest')
plt.xticks(rotation='vertical')
plt.ylabel('Player overall rating')
plt.xlabel('Country')
```
