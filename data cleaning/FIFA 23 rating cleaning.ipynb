```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
rating23 = pd.read_csv('../Datasets/FIFA/FIFA23_rating/player_rating.csv')
```

```python
def drop_1(dataset):
    '''drop columns with more than 200 missing data or more than 200 values = 0'''
    na_s = rating23.isnull().sum()
    col_0 = (rating23 == 0).sum(axis = 0)
    drop_list = [i for i in na_s.index if na_s[i] > 200]
    for i in col_0.index:
        if col_0[i]>200:
            drop_list.append(i)
    dataset.drop(columns = drop_list, inplace = True)
```

```python
drop_1(rating23)
```

```python
new_col = ['fullNameForSearch', 'birthdate', 'preferredposition1label', 'overall_rating','pac','sho','pas','dri','def','phy', 
           'longshots', 'volleys','balance', 'shotpower', 
           'height', 'longpassing', 'standingtackle', 'crossing', 'nationality',
           'curve', 'finishing','slidingtackle','aggression','sprintspeed', 'headingaccuracy', 
           'reactions', 'ranking', 'penalties', 'strength', 'skillmoves', 'stamina', 
           'ballcontrol', 'acceleration', 'weight', 'preferredfoot',
           'vision',  'agility', 'composure', 'jumping', 
           'dribbling', 'interceptions', 'shortpassing']
```

```python
rating23_new = rating23[new_col]
```

```python
rating23_new['birthdate']=rating23_new.apply(lambda x:x['birthdate'].split()[0], axis=1)
```

```python
rating23_new['birthdate'] = rating23_new.apply(lambda x:datetime.strptime(x['birthdate'], '%m/%d/%Y').year, axis =1)
```

```python
rating23_new = rating23_new.rename({'fullNameForSearch':'full_name', 'birthdate':'birthyear', 'preferredposition1label': 'position'}, axis=1)
```
