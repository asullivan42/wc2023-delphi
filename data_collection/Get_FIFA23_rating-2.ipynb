```python
import requests
import pandas as pd
import json
import time
```

```python
url = 'https://ratings-api.ea.com/v2/entities/fifa-23-womens-rating?sort=ranking%3AASC&limit=1200'

r = requests.get(url)

json_data = json.loads(r.text)

data_docs = json_data['docs']
```

```python
pd.set_option('display.max_columns', None)
```

```python
length = len(data_docs)
```

```python
FIFA23_rating = pd.DataFrame.from_dict(data_docs)
```
